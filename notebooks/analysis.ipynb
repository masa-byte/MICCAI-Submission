```python
from torchvision import transforms
import torch
from PIL import Image
import matplotlib.pyplot as plt
import os
import json
from tqdm import tqdm
import pandas as pd

from models import u_net, deeplabv3, fcb_former, fcn, fpn_net, hr_net, link_net
from dataset_loader import ValidationDataset
from solver import Solver
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def load_model(model_name):
    if model_name == "u-net":
        model = u_net.UNet(in_channels=1, out_channels=1)
    elif model_name == "deeplabv3":
        model = deeplabv3.Deeplabv3Plus()
    elif model_name == "fcb-former":
        model = fcb_former.FCBFormer(num_classes=1, in_channels=1)
    elif model_name == "fcn":
        model = fcn.FCN()
    elif model_name == "fpn-net":
        model = fpn_net.FPN(fpn_net.Bottleneck, [2, 2, 2, 2])
    elif model_name == "hr-net":
        model = hr_net.HRNet()
    elif model_name == "link-net":
        model = link_net.LinkNet(n_classes=1)
    else:
        raise ValueError("Invalid model name")
    return model


# Testing model
def load_best_model(name):
    save_path = f"/scratch/project_2011942/best-models/{name}.pth"
    model = load_model("u-net")
    model.load_state_dict(torch.load(save_path, map_location=device))
    model.to(device)
    model.eval()
    return model
```

```python
# Visualize the results
def visualize_comparison(
    image_tensor,
    mask,
    mask_pred1,
    mask_pred2,
    mask_pred3,
    name1,
    name2,
    name3,
    metric,
    contender_name
):
    # Ensure tensors are detached and converted to numpy
    image_np = image_tensor.squeeze(0).numpy()  # Remove the singleton dimension
    mask = mask.squeeze(0).numpy()  # Remove the singleton dimension
    mask_np1 = mask_pred1.squeeze(0).numpy()  # Remove the singleton dimension
    mask_np2 = mask_pred2.squeeze(0).numpy()  # Remove the singleton dimension
    mask_np3 = mask_pred3.squeeze(0).numpy()  # Remove the singleton dimension

    # Convert tensors to PIL images
    image_pil = Image.fromarray((image_np * 255).astype("uint8")).convert(
        "L"
    )  # Scale image to 0-255
    mask_pil = Image.fromarray((mask * 255).astype("uint8")).convert("L")  # Binary mask
    mask_pil1 = Image.fromarray((mask_np1 * 255).astype("uint8")).convert(
        "L"
    )  # Binary mask
    mask_pil2 = Image.fromarray((mask_np2 * 255).astype("uint8")).convert(
        "L"
    )  # Binary mask
    mask_pil3 = Image.fromarray((mask_np3 * 255).astype("uint8")).convert(
        "L"
    )  # Binary mask

    # Plot the images
    fig, axes = plt.subplots(1, 4, figsize=(16, 8))

    # Display scan image with tumor overlay on the right
    axes[0].imshow(image_pil, cmap="gray", extent=[0, 256, 0, 256])  # Define bounds
    axes[0].imshow(
        mask_pil, cmap="jet", alpha=0.5, extent=[0, 256, 0, 256]
    )  # Match bounds
    axes[0].set_xlim(0, 256)  # Explicitly set x-axis limits
    axes[0].set_ylim(0, 256)  # Explicitly set y-axis limits
    axes[0].set_title(f"Ground Truth")

    # Display scan image with tumor overlay on the right
    axes[1].imshow(image_pil, cmap="gray", extent=[0, 256, 0, 256])  # Define bounds
    axes[1].imshow(
        mask_pil1, cmap="jet", alpha=0.5, extent=[0, 256, 0, 256]
    )  # Match bounds
    axes[1].set_xlim(0, 256)  # Explicitly set x-axis limits
    axes[1].set_ylim(0, 256)  # Explicitly set y-axis limits
    axes[1].set_title(f"{name1}")

    # Display scan image with tumor overlay on the right
    axes[2].imshow(image_pil, cmap="gray", extent=[0, 256, 0, 256])  # Define bounds
    axes[2].imshow(
        mask_pil2, cmap="jet", alpha=0.5, extent=[0, 256, 0, 256]
    )  # Match bounds
    axes[2].set_xlim(0, 256)  # Explicitly set x-axis limits
    axes[2].set_ylim(0, 256)  # Explicitly set y-axis limits
    axes[2].set_title(f"{name2}")

    # Display predicted mask on the right
    axes[3].imshow(image_pil, cmap="gray", extent=[0, 256, 0, 256])  # Define bounds
    axes[3].imshow(
        mask_pil3, cmap="jet", alpha=0.5, extent=[0, 256, 0, 256]
    )  # Match bounds
    axes[3].set_xlim(0, 256)  # Explicitly set x-axis limits
    axes[3].set_ylim(0, 256)  # Explicitly set y-axis limits
    axes[3].set_title(f"{name3}")

    # Add a title
    plt.suptitle(f"Comparison of {metric} between {contender_name} and other models")
    
    plt.tight_layout()
    plt.show()
```

```python
def visualize_metrics(scan_metrics_list, scan_types, metric_name, names): 
    num_scans = len(scan_types)
    fig, ax = plt.subplots(num_scans, 1, figsize=(10, 5 * num_scans))  # Adjust height dynamically

    if num_scans == 1:  # If there's only one scan type, `ax` is not a list
        ax = [ax]

    for i, scan_type in enumerate(scan_types):
        values = [scan_metrics[scan_type][metric_name] for scan_metrics in scan_metrics_list]
        ax[i].bar(names, values)
        ax[i].set_title(f"{metric_name} for {scan_type}")
        ax[i].set_ylim(0, 1)
    
    plt.tight_layout()
    plt.show()
    
```

```python
def test(model, name, image, mask):
    mask_pred, dsc, specificity, sensitivity = Solver.test_model(
        model, image, mask, device
    )

    print(name)
    print(f"Unique values in predicted mask: {torch.unique(mask_pred)}")
    print(f"Dice Similarity Coefficient: {dsc}")
    print(f"Specificity: {specificity}")
    print(f"Sensitivity: {sensitivity}")
    print()

    return mask_pred
```

```python
# Load validation dataset
test_paths = []
test_path = "/scratch/project_2011942/datasets/test"
test_paths.extend(os.path.join(test_path, path) for path in os.listdir(test_path))

transformation = transforms.Compose(
    [transforms.Resize((256, 256)), transforms.ToTensor()]
)

validation_dataset = ValidationDataset(test_paths, transformation)
validation_loader = torch.utils.data.DataLoader(
    validation_dataset, batch_size=1, shuffle=False, num_workers=2, prefetch_factor=2
)
```

```python
def get_insights(model, name):
    scan_type_metrics, image_metrics = Solver.analyze_model(
        model, validation_loader, device
    )

    print(f"Model: {name}")
    print(scan_type_metrics)

    return scan_type_metrics, image_metrics
```

```python
name1 = "UNetFocalLoss2025-02-05_07-04-45"
name2 = "UNetBCEWithLogitsLoss2025-02-04_01-30-19"
name3 = "UNetAdaptiveDistributionLoss2025-02-05_14-25-07"

model1 = load_best_model(name1)
model2 = load_best_model(name2)
model3 = load_best_model(name3)

# Split names on the first number
name1 = name1.split("2025")[0]
name2 = name2.split("2025")[0]
name3 = name3.split("2025")[0]
```

```python
scan_metrics1, image_metrics1 = get_insights(model1, name1)
scan_metrics2, image_metrics2 = get_insights(model2, name2)
scan_metrics3, image_metrics3 = get_insights(model3, name3)
```

```python
metrics = {
    "name1": name1,
    "scan_metrics1": scan_metrics1,
    "name2": name2,
    "scan_metrics2": scan_metrics2,
    "name3": name3,
    "scan_metrics3": scan_metrics3
}

with open(f"{model1.__class__.__name__}_metrics.json", "w") as f:
    json.dump(metrics, f, indent=4)  # `indent=4` makes it more readable
```

```python
# Visualize metrics
scan_types = ["T1", "T2", "FLAIR"]
metrics = [scan_metrics1, scan_metrics2, scan_metrics3]
```

```python
visualize_metrics(metrics, scan_types, "specificity", [name1, name2, name3])
```

```python
# Comparison of best vs worst predictions for each model
index = 0
for image_metric in [image_metrics1, image_metrics2, image_metrics3]:
    contender_name = [name1, name2, name3][index]
    index += 1

    for metric in ["minimal_dsc", "maximal_dsc", "minimal_specificity", "maximal_specificity", "minimal_sensitivity", "maximal_sensitivity"]:
        print(f"Metric: {metric}")

        image, mask, metadata = image_metric[metric][1], image_metric[metric][2], image_metric[metric][3]

        image = image.squeeze(0)
        mask = mask.squeeze(0)

        print(f"Metadata: {metadata}")
        print(f"Unique values in mask: {torch.unique(mask)}")
        print()

        pred1 = test(model1, name1, image, mask)
        pred2 = test(model2, name2, image, mask)
        pred3 = test(model3, name3, image, mask)
        visualize_comparison(image, mask, pred1, pred2, pred3, name1, name2, name3, metric, contender_name)
```

```python
# Calculating how many scans are a false positive/negative for each model
false_positives1, false_negatives1 = 0, 0
false_positives2, false_negatives2 = 0, 0
false_positives3, false_negatives3 = 0, 0

# Assumes model is already in evaluation mode
# Assumes batch size of 1
for batch in tqdm(validation_loader, desc="Analyzing FP and FN"):
    image, mask, metadata = batch

    image = image.to(device)
    mask = mask.to(device)

    with torch.no_grad():
        mask_pred = model1(image)
        # Ensure predicted mask is binary
        binary = (mask_pred > 0).float()

        if binary.sum() == 0 and mask.sum() > 0:
            false_negatives1 += 1
        elif binary.sum() > 0 and mask.sum() == 0:
            false_positives1 += 1

    with torch.no_grad():
        mask_pred = model2(image)
        # Ensure predicted mask is binary
        binary = (mask_pred > 0).float()

        if binary.sum() == 0 and mask.sum() > 0:
            false_negatives2 += 1
        elif binary.sum() > 0 and mask.sum() == 0:
            false_positives2 += 1

    with torch.no_grad():
        mask_pred = model3(image)
        # Ensure predicted mask is binary
        binary = (mask_pred > 0).float()

        if binary.sum() == 0 and mask.sum() > 0:
            false_negatives3 += 1
        elif binary.sum() > 0 and mask.sum() == 0:
            false_positives3 += 1     
```

```python
# Plot the results
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].bar(["False Positives", "False Negatives"], [false_positives1, false_negatives1])
ax[0].set_title(name1)

ax[1].bar(["False Positives", "False Negatives"], [false_positives2, false_negatives2])
ax[1].set_title(name2)

ax[2].bar(["False Positives", "False Negatives"], [false_positives3, false_negatives3])
ax[2].set_title(name3)

plt.tight_layout()
plt.show()
```

```python
# Make a dataframe
data = {
    "Model": [name1, name2, name3],
    "False Positives": [false_positives1, false_positives2, false_positives3],
    "False Negatives": [false_negatives1, false_negatives2, false_negatives3],
    "PercentageFP": [false_positives1 / len(validation_loader) * 100, false_positives2 / len(validation_loader) * 100, false_positives3 / len(validation_loader) * 100],
    "PercentageFN": [false_negatives1 / len(validation_loader) * 100, false_negatives2 / len(validation_loader) * 100, false_negatives3 / len(validation_loader) * 100]
}

df = pd.DataFrame(data)
df.to_csv(f"{model1.__class__.__name__}_false_positives_negatives.csv", index=False)

df
```
